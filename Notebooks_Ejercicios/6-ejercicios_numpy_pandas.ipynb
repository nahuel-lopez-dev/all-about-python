Pista: Usar funcion .text.split()


```python
def contar_palabras(cadena):
    palabras = cadena.split()
    frecuencias = {}

    for palabra in palabras:
        # Eliminar puntuaciones al final de la palabra
        palabra = palabra.rstrip('.,?!')

        # Convertir la palabra a minúsculas para evitar distinciones entre mayúsculas y minúsculas
        palabra = palabra.lower()

        if palabra in frecuencias:
            frecuencias[palabra] += 1
        else:
            frecuencias[palabra] = 1

    return frecuencias

def palabra_mas_repetida(frecuencias):
    if not frecuencias:
        return None

    palabra_max = max(frecuencias, key=frecuencias.get)
    frecuencia_max = frecuencias[palabra_max]

    return (palabra_max, frecuencia_max)

# Solicitar al usuario una cadena de caracteres
cadena_input = input("Ingrese una cadena de caracteres: ")

# Obtener el diccionario de frecuencias
frecuencias_dict = contar_palabras(cadena_input)

# Mostrar el diccionario de frecuencias
print("Diccionario de frecuencias:")
print(frecuencias_dict)

# Obtener la palabra más repetida y su frecuencia
palabra_repetida_info = palabra_mas_repetida(frecuencias_dict)

# Mostrar la palabra más repetida y su frecuencia
if palabra_repetida_info:
    print(f"Palabra más repetida: '{palabra_repetida_info[0]}' con frecuencia {palabra_repetida_info[1]}")
else:
    print("No hay palabras para analizar.")
```

### `Ejercicio 2 - Coma`
Cuando Takahashi escribe un número entero, utiliza una coma cada tercer dígito desde la derecha. Por ejemplo, 1234567, se escribe como 1,234,567 y 777 se escribe como 777
¿Cuántas comas se utilizarán en total cuando escriba cada número entero del 1 al N una vez?


```python
def contar_comas_hasta_N(N):
    comas_totales = 0
    i = 1

    while i <= N:
        comas_totales += min(N - i + 1, 10**((len(str(i)) - 1)//3))
        i *= 10

    return comas_totales

# Solicitar al usuario el valor de N
N = int(input("Ingrese un número entero N: "))

# Calcular y mostrar la cantidad total de comas
resultado = contar_comas_hasta_N(N)
print(f"La cantidad total de comas al escribir los números del 1 al {N} es: {resultado}")
```

### `Ejercicio 3 en clase - Búsqueda Binaria`

```python
def binary_search(arr, target):

  inicio, fin = 0, len(arr) - 1
  # print(inicio, fin)

  while inicio<=fin:
    medio =  (inicio + fin)//2
    if arr[medio] == target:
      return medio
    elif arr[medio] < target:
      inicio = medio+1
    else:
      fin = medio-1

  return -1

arr = [0,4,7,10,14,23,45,47,53]
target = 53

resultado = binary_search(arr,target)
if resultado != -1 :
  print("Se encontro el valor en el indice", resultado)
else:
  print("No se encontro el valor")
```

### `Ejercicio opcional 3`
 Escribí una función recursiva is_poten(N, b) que reciba 2 enteros, N y b, y
devuelva True si N es potencia de b y False en caso contrario.
- is_poten(8, 2) -> True
- is_poten(64, 4) -> True
- is_poten(70, 10) -> False
- is_poten(1, 2) -> True


```python
def is_poten(N, b):
    # Caso base: N es igual a 1, ya que cualquier número elevado a 0 es 1
    if N == 1:
        return True
    # Caso base: N es menor que b, ya que cualquier número elevado a 0 es 1
    elif N < b:
        return False
    else:
        # Llamada recursiva con N dividido por b
        return is_poten(N // b, b)

# Ejemplos de uso
print(is_poten(8, 2))    # True
print(is_poten(64, 4))   # True
print(is_poten(70, 10))  # False
print(is_poten(1, 2))    # True
```

### `Ejercicio Calificado 2 - Suma con recursión`
Dada una matriz de números enteros, encuentre la suma de los elementos de
la matriz usando recursividad.


```python
def suma_matriz(matriz):
    # Caso base: matriz vacía
    if not matriz:
        return 0

    # Sumar el primer elemento de la primera fila y la suma de los elementos restantes
    return sum(matriz[0]) + suma_matriz(matriz[1:])

# Ejemplo de uso
matriz_ejemplo = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

resultado = suma_matriz(matriz_ejemplo)
print(f"La suma de los elementos de la matriz es: {resultado}")
```

### `Ejercicio Calificado 3 - Search`
 Crear una funcion. Se le pasara como argumento
una lista. La salida de la funcion debe ser la lista
ordenada. La logica del programa debe ser la
mostrada en la imagen.
 - Input : [2 , 13 , 8 , 0]
 - Output : [0 , 2 , 8 , 13]


```python
def ordenar_lista(lista):
    n = len(lista)

    for i in range(n):
        for j in range(0, n-i-1):
            if lista[j] > lista[j+1]:
                lista[j], lista[j+1] = lista[j+1], lista[j]

    return lista

# Ejemplo de uso
input_lista = [2, 13, 8, 0]
output_lista = ordenar_lista(input_lista)

print("Input:", input_lista)
print("Output:", output_lista)
```

maxima C de la bolsa. Debe de agregar los utils
utilizando un bucle for

```python
def llenar_bolsa(capacidad_maxima, pesos):
    mejor_combinacion = []
    mejor_utilidad = 0

    for i in range(len(pesos) + 1):
        for combinacion in obtener_combinaciones(pesos, i):
            peso_total = sum(combinacion)
            if peso_total <= capacidad_maxima:
                utilidad_total = sum(obtener_utilidades(combinacion))
                if utilidad_total > mejor_utilidad:
                    mejor_combinacion = combinacion
                    mejor_utilidad = utilidad_total

    return mejor_combinacion

def obtener_combinaciones(items, n):
    if n == 0:
        return [[]]

    combinaciones = []
    for i, item in enumerate(items):
        restantes = items[i+1:]
        for sub_combinacion in obtener_combinaciones(restantes, n-1):
            combinaciones.append([item] + sub_combinacion)

    return combinaciones

def obtener_utilidades(combinacion):
    return [utilidad(peso) for peso in combinacion]

def utilidad(peso):
    # Esta función puede ser personalizada según las utilidades de los objetos
    # En este ejemplo, simplemente retornamos el peso como utilidad
    return peso

# Ejemplo de uso
capacidad_maxima = 47
pesos = [7, 3, 0.5]

mejor_combinacion = llenar_bolsa(capacidad_maxima, pesos)

print(f"Mejor combinación de elementos para una capacidad de {capacidad_maxima} Kg:")
print(mejor_combinacion)
```

### `Ejercicio 1`
Esta función toma dos matrices 2D 𝐴 y 𝐵 y devuelve la suma de las columnas
 de A multiplicada por la suma de todos los elementos de 𝐵, i.e., a scalar.




```python
import numpy as np

def suma_columnas_mult_suma_elementos(A, B):
    # para sumar de las columnas de A
    suma_columnas_A = np.sum(A, axis=0)

    # para sumar de todos los elementos de B
    suma_elementos_B = np.sum(B)

    # para multiplicar la suma de columnas de A por la suma de elementos de B
    resultado = np.dot(suma_columnas_A, suma_elementos_B)

    return resultado

# Ejemplo de uso
A = np.array([[1, 1],
              [1, 1]])

B = np.array([[1, 2, 3],
              [1, 2, 3]])

resultado = suma_columnas_mult_suma_elementos(A, B)
print("Resultado:", resultado)
```

### `Ejercicio 2`
 Esta función toma una matriz cuadrada 𝐶 y devuelve una matriz 2D que consta
 de un 𝐶 aplanado con el índice de cada elemento adjunto a este tensor en la
 dimensión de la fila.

```python
import numpy as np

def aplanar_con_indices(C):
    #  para aplanar la matriz C
    C_aplanada = C.flatten()

    # para crear un arreglo con índices
    indices = np.arange(C.size)

    # para concatenar los índices como una nueva columna
    resultado = np.column_stack((indices, C_aplanada))

    return resultado

# Ejemplo de uso
C = np.array([[2, 3],
              [-1, 10]])

resultado = aplanar_con_indices(C)
print("Resultado:")
print(resultado)
```

### `Ejercicio 3`
 Crear una funcion que reciba de argumentos dos vectores A y B , y ejecute la
 multiplicacion matricial de A y B usando funciones integradas en la librería
 numpy.
 Crear otra funcion que ejecute la multiplicacion matricial iterando sobre cada
 uno de los valores de ambos vectores (usando for) .
 Comparar el tiempo de ejecución.


```python
import numpy as np
import time

def multiplicacion_matricial_numpy(A, B):
    return np.dot(A, B)

def multiplicacion_matricial_iterativa(A, B):
    resultado = 0
    for a, b in zip(A, B):
        resultado += a * b
    return resultado

# Ejemplo de uso
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

# Multiplicación matricial con Numpy
inicio_numpy = time.time()
resultado_numpy = multiplicacion_matricial_numpy(A, B)
fin_numpy = time.time()

# Multiplicación matricial iterativa
inicio_iterativo = time.time()
resultado_iterativo = multiplicacion_matricial_iterativa(A, B)
fin_iterativo = time.time()

# Mostrar resultados y tiempos de ejecución
print("Resultado con NumPy:", resultado_numpy)
print("Resultado iterativo:", resultado_iterativo)

tiempo_numpy = fin_numpy - inicio_numpy
tiempo_iterativo = fin_iterativo - inicio_iterativo

print("Tiempo de ejecución con NumPy:", tiempo_numpy)
print("Tiempo de ejecución iterativo:", tiempo_iterativo)
```

 Esta función toma dos matrices 2D 𝐷 y 𝐸. Si las dimensiones lo permiten, esta
 función devuelve la suma por elementos de 𝐷 en forma de 𝐸 y 𝐷; de lo
 contrario, esta función devuelve una matriz 1D que es la concatenación de las
 dos matrices.
 > Pista: np.size() es una manera fácil de encontrar el número de elementos en una matriz.



```python
import numpy as np

def suma_por_elementos_o_concatena(D, E):
    # Verificar si el número total de elementos permite la suma por elementos
    if D.size == E.size:
        resultado = D + E.reshape(D.shape)
    else:
        # Concatenar las matrices en una matriz 1D
        resultado = np.concatenate((D.flatten(), E.flatten()))

    return resultado

# Ejemplo de uso
D = np.array([[1, -1],
              [-1, 3]])

E = np.array([2, 3, 0, 2])
# E = np.array([2, 3, 0])

resultado = suma_por_elementos_o_concatena(D, E)
print("Resultado:")
print(resultado)

# Resultados que debe arrojar:

# con:
# D = np.array([[1, -1],
#               [-1, 3]])
# y
# E = np.array([2, 3, 0, 2])
# el resultado es:
# Resultado 1
#  [3,  2]
#  [-1, 5]

# y si E es:
# E = np.array([2, 3, 0])

# Resultado 2
# [ 1 -1 -1  3  2  3  0]
```

### `Ejercicio Calificado 2`
 Se solicita implementar una función llamada crear_cuadrado que reciba como
 entrada una coordenada (x, y), un numero entero M y otro square_size. La
 función debe crear un cuadrado utilizando un arreglo NumPy, donde la
 coordenada de la esquina superior izquierda del cuadrado sea (x, y), la
 longitud de la imagen(arreglo) es M y el tamaño del cuadro es square_size.
 Si el cuadrado no puede ser creado debido a que se encuentra en uno de sus
 bordes y no hay suficiente espacio, la función debe imprimir el mensaje 'No se
 puede crear el cuadrado en la posición proporcionada debido a la falta de
 espacio en los bordes' y retornar None.
 La función debe retornar el arreglo NumPy que representa el cuadrado si se
 pudo crear exitosamente.

 > Nota: los pixeles del cuadrado deben tener valor 255 y los pixeles del fondo
 deben tener valor 0


```python
import numpy as np

def crear_cuadrado(x, y, M, square_size):
    # Verificar si hay suficiente espacio para el cuadrado
    if x < 0 or y < 0 or x + square_size > M or y + square_size > M:
        print('No se puede crear el cuadrado en la posición proporcionada debido a la falta de espacio en los bordes')
        return None

    # Crear un arreglo NumPy de MxM con valores 0 (fondo)
    imagen = np.zeros((M, M), dtype=np.uint8)

    # Establecer los valores del cuadrado en 255
    imagen[x:x+square_size, y:y+square_size] = 255

    return imagen

# Ejemplo de uso
x = 1
y = 1
M = 10
square_size = 2

resultado = crear_cuadrado(x, y, M, square_size)

if resultado is not None:
    print(resultado)
```
