### `Ejemplo de programa m√°s avanzado`

Veamos un ejemplo de un peque√±o programa. El mismo est√° hecho para resolver el siguiente problema:

> Una ma√±ana se pone un billete en la vereda al lado del obelisco porte√±o. A partir de ah√≠, cada d√≠a se duplica la cantidad de billetes que hay, apil√°ndolos prolijamente. ¬øCu√°ntos d√≠as pasan antes de que la pila de billetes sea m√°s alta que el obelisco?


No te preocupes ahora por entender todo lo que hace. Con este programa como base, veamos los elementos b√°sicos de un c√≥digo de Python:

````python
# un billete mide 0.11 milimetros de espesor
grosor_billete = 0.11 / 1000   # espesor de un billete en metros
altura_obelisco = 67.5         # altura en metros
num_billetes = 1
dia = 1

while num_billetes * grosor_billete <= altura_obelisco:
    # mientras la altura de la pila de billetes sea menor o igual a la altura del obelisco, 
    # ejecutamos este bloque
    print(dia, num_billetes, num_billetes * grosor_billete)
    dia = dia + 1
    num_billetes = num_billetes * 2

print('Cantidad de d√≠as', dia)
print('Cantidad de billetes', num_billetes)
print('Altura final', num_billetes * grosor_billete)
````

### `Booleanos (bool)`

Las variables booleanas son muy sencillas, solo pueden tomar dos valores: `True` o `False` (verdadero o falso). En Python estas variables son tambi√©n num√©ricas porque **True** equivale a **1** y **False** a **0**.

````python
a = True  # verdadero
b = False # falso
````

Como veremos m√°s adelante, los boolenos sirven para trabajar con expresiones l√≥gicas. Por ejemplo, para ver si un valor es mayor que otro, usamos el operador `>`. El resultado es un valor booleano:

````python
12 > 12
````

Para conocer el **tipo** de variable en Python, podemos usar la funci√≥n `type`, pasandole como argumento una variable. As√≠:

Hay adem√°s tres operadores que sirven para trabajar con valores l√≥gicos (booleanos): `and`, `or` y `not` ('y', 'o' y 'no'). El `and` compara dos booleanos, determinando si *ambos* son verdaderos o no:

````python
print(True and True)   # verdadero y verdadero?
print(False and True)  # falso y verdadero?
print(False and False) # falso y falso?
````

El `or` determina si *alguno* de los dos booleanos es verdadero:

````python
print(True or True)   # verdadero o verdadero?
print(False or True)  # falso o verdadero?
print(False or False) # falso o falso?
````

Finalmente, el `not` niega un valor booleano (toma el opuesto)

````python
print(not False)
print(False or (not True))
````

### `C√≥digos de escape`

Los c√≥digos de escape (*`escape codes`*) son expresiones que comienzan con una barra invertida, `\` y se usan para representar caracteres que no pueden ser f√°cilmente tipeados directamente con el teclado. Estos son algunos c√≥digos de escape usuales:

````python
# \         para no tener en cuenta el signo que sigue
# Ej: print("La ciudad que conoc√≠ es \"maravillosa\"") y as√≠ va a tomar las dobles comillas de "maravillosa"
# otra alternativa, usando simples y dobles comillas: print('La ciudad que conoc√≠ es "maravillosa"')

# otros escape codes:
# '\n'      Avanzar una l√≠nea, pega un salto a la siguiente l√≠nea
# '\r'      Retorno de carro
# '\t'      Tabulador
# '\''      Comilla literal
# '\"'      Comilla doble literal
# '\\'      Barra invertida literal
````

### `Indexaci√≥n de cadenas`

Las cadenas funcionan como los vectores en matem√°tica o los arrays. A pesar de que se especifican todos juntos (entre comillas), podemos acceder a caracteres individuales por su posici√≥n en la cadena. El √≠ndice comienza a contar en cero. Los √≠ndices negativos se usan para especificar una posici√≥n respecto al final de la cadena.

En Python es f√°cil tomar una porci√≥n de una cadena de caracteres. Esto se llama tomar una rebanda (o en ingl√©s *slice*). Para eso se utiliza un rango de posiciones (√≠ndices), usando el operador `:`

### `Ordenar listas`

### M√©todo `sort` (modifica la lista original)
Las listas pueden ser ordenadas "in-place", es decir, sin asignarlas a nuevas variables.

````python
s = [10, 1, 7, 3]
s.sort()                    # [1, 3, 7, 10]

# Orden inverso
s = [10, 1, 7, 3]
s.sort(reverse=True)        # [10, 7, 3, 1]

# Funciona con cualquier tipo de datos que tengan orden
s = ['foo', 'bar', 'spam']
s.sort()                    # ['bar', 'foo', 'spam']
````

### Funci√≥n `sorted` (No modifica la lista origina, retorna otra)
`sorted()` retorna nueva lista ordenada, en lugar de ordenar la misma, y esta lista se puede guardar en una variable:

````python
t = sorted(s)               # s queda igual, t guarda los valores ordenados
````
