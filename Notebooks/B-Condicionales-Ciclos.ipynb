### **`Listas`** 

#### Creación 
Supongamos que tenemos que ir a la fruteria y comprar: manzanas, peras, bananas y naranjas. Podemos crear una variable para cada caso:


```python
fruta1 = 'manzanas'
fruta2 = 'peras'
fruta3 = 'bananas'
fruta4 = 'naranjas'
```


La computadora no sabe que son frutas, ella interpreta las 4 variables independientemente, y podemos acceder a cada una preguntando por el valor de ella. 


```python
print(fruta1)
```

Pero, ¿cómo podemos agruparlas de forma más eficiente? Por ejemplo, aguparlas de forma que sepamos que todas son frutas. Una forma es hacerlo mediante la creación de una lista con todas las frutas en ella. 

Las listas son el tipo de datos primitivo de Python para guardar colecciones ordenadas de valores. Para definir una lista se usan corchetes: 

```python
frutas = ['manzanas', 'peras', 'bananas', 'naranjas']
```

```python
frutas2 = [fruta1, fruta2, fruta3, fruta4]
```

```python
print(frutas)
```

¿Cuál es el primer elemento? ¿Cómo nos referimos a él?

```python
print(frutas[0])  # imprimimos el primer elemento de la lista frutas
```

Tambien podemos empezar a contar de atrás para adelante. El último elemento de la lista tiene indice -1, el anteúltimo -2 y así por delante.

## Condicionales

### `if`
El comando `if` ("si" en español) es usado para ejecutar un condicional:

```python
a = 7
b = 5

if a > b:
    print('Gana a')
```

El código dentro del bloque indentado del `if` sólo se va a ejecutar si se cumple la condición, en este caso `a > b`. Si queremos que nuestro código haga algo distinto si no se cumple la condición, podemos usar un `else` ("si no"):

### `if` con `else`

```python
a = 7
b = 5

if a > b:
    print('Gana a')
else:
    print('Gana b')
```

El código anterior se lee como: "si a es mayor que b, imprimí `'Gana a'`. Si no, imprimí `'Gana b'`".  
Se pueden verificar condiciones mutuamente excluyentes agregando condiciones extras con `elif`.

```python
calificacion = 8

if calificacion > 7:
    print('Felicidades, aprobaste!')
else:
    print('Reprobaste la materia.')
```

### `Ternario` para reducir líneas de código con if y else

```python
calificacion = 9
    
nota = 'Felicidades aprobaste la materia' if calificacion >= 7 else 'Desaprobaste la materia'
print(f'{nota} con {calificacion}')
```

```python
calificacion = 10

color = 'verde' if calificacion >= 7 else 'Rojo'

print(f'El color es "{color}" porque su calificación es {calificacion}!')
```

### `Comandos if, else y elif`

```python
a = 3
b = 5

if a > b:
    print('Gana a')
elif a == b:
    print('Empate!')
else:
    print('Gana b')
```

El comando `elif` viene de *else* e *if* y puede traducirse como "*si no se da la condición del if anterior, verificá si se da la siguiente*".
Las condiciones siempre van a estar dadas por un valor booleano. Por supuesto que podemos combinar varias condiciones usando los operadores entre booleanos que vimos antes (`and`, `or`, `not`):

```python
a = 1
b = 2
c = 3

if a > b and a > c:
  print('a es el más grande')
```

No hace nada porque la condición no se cumple es False.
Otro ejemplo puede ser:

```python
a = 1
b = 2
c = 3

if a < b or a < c:
  print('a no es el más grande')
```

```python
calificacion = 10

if calificacion == 10:
   print('Felicidades, aprobaste la materia con una calificación perfecta.')
elif calificacion == 9 or calificacion == 8:
    print('Felicidades, aprobate la materia.')
elif calificacion == 7 or calificacion == 6:
    print('Aprobaste la materia.')
else:
    print('Reprobaste la materia.')
```

## Ciclos, bucles, loops o iteradores
Tienen diferentes nombres pero hacen lo mismo: iteran. Nos permiten ahorrar lineas de programacion y ejecutar la misma secuencia de comandos varias veces. 



## Ciclo `for`

En Python, los bucles se pueden programar de diferentes maneras. La más común es el bucle `for` (que podemos traducir al español con `para`), que se utiliza junto con objetos iterables, como las listas. La sintaxis básica es:


```python
for i in [0,1,2]:
  print(i)
```

### Iterando sobre una  lista de palabras:

```python
usuarios = ['usuario1', 'usuario2', 'usuario3', 'usuario4']

for usuario in usuarios:
    print(usuario)
```

```python
frutas = ['manzanas', 'peras', 'bananas', 'naranjas']
for fruta in frutas:
  print(fruta)
```

En cada iteración del bucle, vamos asignando ordenadamente un nuevo elemento a la variable `usuario` en el primer caso y `fruta` en el segundo. El valor previo es sobreescrito, y al final del bucle retiene su último valor.

### Iterando sobre un string:

```python
for char in 'ultimate python':
    print(char, end='') 
# agregando end='' para retornar el resultado de forma horizontal
# por defecto lo devuelve vertical, iterando letra por letra y pasando a la línea de abajo 
```

### La función `range`
El bucle `for` itera sobre los elementos de la lista suministrada, y ejecuta el bloque contenedor una vez por cada elemento. En el bucle `for` se puede utilizar cualquier tipo de lista. Por ejemplo:

```python
for i in range(3):   # range comienza en 0   
  print(i)
```

*Atención, `range(3)` no incluye al 3!*

`range(3)` crea una lista que contiene los números naturales < 3, es decir `[0, 1, 2]`, la misma lista que le dimos anteriormente.

### range admite parámetros: `range(start, end, skips)`

```python
for valor in range(5, 21, 3):
   print(valor)
```

### `El valor de skip por defecto es 1`

```python
for x in range(-3,3):     # números enteros desde el -3 y menores que 3!
    print(x)
```

### La función `enumerate`

```python
numeros = [10, 20, 30, 40, 50]

for indice, numero in enumerate(numeros):
    print(f'El índice es {indice} y el número es {numero}')
```

### `break` y `continue`

#### Usando `break`

```python
titulo_curso = 'Curso profesional de Python'

for caracter in titulo_curso:

    if caracter == 'P':
        break

    print(caracter, end=' ')
```

#### Usando `continue`

```python
titulo_curso = 'Curso profesional de Python'

for caracter in titulo_curso:

    if caracter == 'P':
        continue

    print(caracter, end='')
```

## Ciclo `while`

El comando `while` (que en español significa `mientras`) ejecuta un ciclo mientras que cierta condición se cumpla. Cuando la condición no se cumple más, el ciclo es finalizado.

```python
i = 0               # defino la variable i

while i < 5:        # creo la condición para hacer el ciclo
    print(i)
    i = i + 1
    
print("listo! ya sali del ciclo")       # una vez que la condición no se cumple más, salimos del ciclo
```

```python
numero = 1

while numero < 100:
    print(numero)
    numero *= 2
```

```python
comando = ''

while comando.lower() != 'salir':
    comando = input("$ ")
    print(comando)
```

Los comandos indentados debajo del `while` se van a a ejecutar mientras que la expresión luego del `while` sea verdadera (`True`).



### `while True`
En Python no hay un `do ... while` como en otros lenguajes, pero en su lugar se suele utilizar el `while True:`

```python
while True:
    comando = input("$ ")
    print(comando)
    if comando.lower() == "salir":
        break
    
```

### `while con else`

```python
numero = 123456789
contador_digitos = 0

while numero >= 1:
    # contador_digitos = contador_digitos + 1
    contador_digitos += 1

    numero = numero / 10
else:
    print(contador_digitos)
```

#### Ejemplo con while

Volvamos al ejemplo de un programa mostrado en la primera clase. 

> Una mañana se pone un billete en la vereda al lado del obelisco porteño. A partir de ahí, cada día se duplica la cantidad de billetes que hay, apilándolos prolijamente. ¿Cuántos días pasan antes de que la pila de billetes sea más alta que el obelisco?

```python
# obelisco.py
grosor_billete = 0.11 * 0.001  # grosor de un billete en metros
altura_obelisco = 67.5         # altura en metros
num_billetes = 1
dia = 1

while altura_obelisco >= num_billetes * grosor_billete :
    print(dia, num_billetes, num_billetes * grosor_billete)
    dia = dia + 1
    num_billetes = num_billetes * 2

print('Cantidad de días', dia)
print('Cantidad de billetes', num_billetes)
print('Altura final', num_billetes * grosor_billete)
```

### `Indentación`

La indentación se usa para marcar grupos de comandos que van juntos. Considerá el ejemplo del programa anterior:

```python
while num_billetes * grosor_billete <= altura_obelisco:
    print(dia, num_billetes, num_billetes * grosor_billete)
    dia = dia + 1
    num_billetes = num_billetes * 2

print('Cantidad de días', dia)
```

La indentación agrupa los comandos siguientes como las operaciones a repetir:

```python
    print('Días:', dia, 'Números de billetes:', num_billetes, 'Grosor acumulado por los billetes:', num_billetes * grosor_billete)
    dia = dia + 1
    num_billetes = num_billetes * 2
```

Como el comando `print()` al final no está indentado, no pertenece al ciclo. La línea en blanco que dejamos entre ambos solo está para facilitar la lectura del código y no afecta la ejecución.

___Algunas recomendaciones sobre cómo indentar___:


*   Usar espacios y no el tabulador (ah no ser que la tabulación sean 4 espacios).
*   Usar 4 espacios por cada nivel (aunque en notebooks suelen usarse 2, `el pep 8` recomienda 4).
*   Usar editores de textos que entiendan que estás escribiendo en Python.

El único requisito verdadero del intérprete de Python es que la indentación dentro de un mismo bloque sea consistente. Por ejemplo, esto es un error:

```python
while num_billetes * grosor_billete <= altura_obelisco:
    print(dia, num_billetes, num_billetes * grosor_billete)
        dia = dia + 1 # ERROR
    num_billetes = num_billetes * 2
```

## Más sobre iteraciones

### Comando break

Podemos usar el comando `break` para interrumpir y terminar el bucle. Al ejecutarlo, el programa sale del bucle y continúa con los comandos que siguen a continuación. Si el bucle está dentro de otro bucle, sólo termina el primero.

```python
numeros = [1, 2, 3, 4, 5]

for i in numeros:
  print(i)
  if i == 3:
    break

print('bucle finalizado','ultimo valor de i',i)
```

### Comando continue

Otro comando útil en bucles es el `continue`. Lo que hace es saltear la actual iteración y pasar inmediatamente a la siguiente. Es útil cuando no nos interesa procesar ciertos elementos.

```python
numeros = [1, -5, 2, -8, 10]

# imprimo los cuadrados de los números positivos
for i in numeros:
  if i <= 0:
    continue
  print(i**2)
```

### Iterando sobre enteros

Para iterar sobre, se suele usar la función `range()`. La sintaxis es `range([inicio,] fin [,paso])`, donde es opcional indicar el número de inicio (por defecto es 0) y el paso (por defecto es 1). El valor de finalización no es incluído, como en las rebanadas.

```python
print('i:')
for i in range(10):
    print(i, end=' ')      # i = 0,1,...,9

print('\nj:')
for j in range(10,20):
    print(j, end=' ')      # j = 10,11,..., 19

print('\nk:')
for k in range(10,20,2):
    print(k, end=' ')      # k = 10,12,...,18
    # Va contando de a 2
```

La función `range()` no almacena todos los valores a la vez, sino lo que los va generando a medida que son necesarios. Muchas veces es tentador usar `range(len(secuencia))` para iterar sobre los índices de una secuencia, pero para eso es mejor el comando `enumerate()`.

### Función enumerate

La función `enumerate()` (enumerar) nos permite agregar un contador a la iteración. Suele ser sumamente útil si, por ejemplo, por alguna razón necesitamos los índices de los elementos.

```python
nombres = ['Juan', 'Elsa', 'Ingrid', 'Carlos', 'Federico']

for i, nombre in enumerate(nombres):
  print(i, nombre)
  # 0 Juan
  # 1 Elsa
  # etc.
```

La forma más general de enumerate es `enumerate(secuencia [, inicio = 0])`, donde el valor de inicio de contador es opcional.

## Secuencias

Las secuenias son una colección ordenada de elementos. Las secuencias poseen indexación y tienen longitud. Vimos tres tipos de secuencias: **cadenas**, **listas** y **tuplas** (que vamos a ver en la próxima semana).

```python
a = 'alumnos'               # cadena
b = [1, 2, 3, 4]             # lista

# Los elementos están ordenados y poseen un índice
a[0]                      # 'a'
b[-1]                     # 4

# Longitudes de secuencias
len(a)                    # 7
len(b)                    # 4
```

Las secuencias pueden ser replicadas y también concatenadas con otras secuencias del mismo tipo:

```python
# Replicación
a*3            # 'alumnosalumnosalumnos'
b*2            # [1, 2, 3, 4, 1, 2, 3, 4] 

# Concatenación
b2 = [5, 6]
b + b2         # [1, 2, 3, 4, 5, 6]
```

También admiten algunas funciones básicas que las reducen a un sólo número, como `max()`, `min()` y `sum()`:

```python
l = [1, 2, 3, 4, 5]
max(l)               # 5
min(l)               # 1
sum(l)               # 15

print(max(l))
print(min(l))
print(sum(l))

# Algunas funcionan incluso con cadenas
t = ['holaaaaa', 'todos'] # ordena alfabéticamente, no cantidad de caracteres
min(t)               # 'hola' ¿Qué estará haciendo?
```

### Slicing

Ya vimos el rebanamiento (slicing) en el caso de listas y cadenas, pero aplican a todo tipo de secuencias. Recordamos que la sintaxis es `s[principio:fin]`, donde el índice final es *excluyente*, el elemento `s[fin]` no queda incluído. Si no se especifican los índices, se toma el inicio/final de la secuencia, dependiendo del caso.

```python
a = [0,1,2,3,4,5,6,7,8]

a[2:5]    # [2,3,4]
a[-5:]    # [4,5,6,7,8]
a[:3]     # [0,1,2]

print(a[2:5]) # contando desde el principio, desde el índice 2 incluido hasta el 5 sin incluirlo
print(a[-5:]) # contando desde el final, desde el 5 elemento incluido hasta el final   
print(a[:3]) # contando desde el principio, hasta el 3 sin incluir el 3
```

Las **listas**, al ser mutables, admiten el borrado y reasignación de rebanadas:

```python
# Reasignación
a = [0,1,2,3,4,5,6,7,8]
a[2:4] = [10,11,12]       # [0,1,10,11,12,4,5,6,7,8]

print(a)

# Borrado
a = [20,21,22,23,24,25,26,27,28]
del a[2:4]                # [20,21,24,25,26,27,28]

print(a)
```

(¡Al reasignar una rebanada de una lista no hace falta que tenga la misma longitud!)

### Ejercicio 7 - (opcional)

Supongamos que tenemos una lista de números enteros, y queremos encontrar todas las subsecuencias de la lista que sumen un número objetivo dado. Para hacerlo, vamos a crear una función que tome como argumentos la lista de números y el número objetivo, y devuelva una lista con todas las subsecuencias que suman el número objetivo. 

*Ayuda:* Por ejemplo, si le damos a tu programa la lista de números [1, 2, 3, 4, 5] y queremos encontrar todas las subsecuencias que suman `7` (número objetivo dado), la función debería devolver las siguientes listas [2, 5], [3, 4], [1, 2, 4].

```python
numeros = [1, 2, 3, 4, 5]
combinaciones = [[]]
resultados = []
objetivo = 7

for n in numeros:
    for i in range(len(combinaciones)):
        nueva = combinaciones[i] + [n]
        combinaciones.append(nueva)
        if sum(nueva) == objetivo:
            resultados.append(nueva)

print (f'Resultado: {resultados}')
```
